```python
import math

# Definindo limites de busca
max_val = 10**6  # Para manter o tempo de execução razoável
tolerancia = 0.001  # 1% de tolerância nas proporções

# Função para checar proporções aproximadas
def checa_paralelepipedo(a, b, c):
    # Diagonais
    diag_menor = math.sqrt(a**2 + b**2)
    diag_maior = math.sqrt(a**2 + b**2 + c**2)

    # Condição da diagonal: maior ≈ 5/3 da menor
    if abs(diag_maior/(diag_menor) - 5/3) > tolerancia:
        return False

    # Áreas dos lados
    areas = [a*b, a*c, b*c]
    areas.sort()
    segunda_maior_area = areas[1]

    # Triângulo retângulo associado ao meio círculo (aprox.)
    tri_area = (b*c)/2  # aproximação
    if abs(tri_area - segunda_maior_area) / segunda_maior_area > tolerancia:
        return False

    # Diagonal da esfera inscrita no cubo (aprox)
    diagonal_cubo = math.sqrt(a**2 + b**2 + c**2)
    # Checa múltiplo de pi/15
    if abs((diagonal_cubo/(math.pi/15)) - round(diagonal_cubo/(math.pi/15))) > tolerancia*100:
        return False

    return True

# Busca de exemplo
encontrado = False
for a in range(1, 500):
    for b in range(a, 500):
        for c in range(b, 500):
            if checa_paralelepipedo(a, b, c):
                print(f"Paralelepípedo aproximado encontrado: a={a}, b={b}, c={c}")
```
