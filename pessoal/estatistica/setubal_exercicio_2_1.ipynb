```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exercício 2.1

# Num lançamento de 3 moedas equilibradas seja X a variável aleatória que representa o número de caras saídas. Escreva a função de probabilidade de X.
```

```python
n, p = 3, 0.5
x = np.arange(0, n + 1)

fp = stats.binom.pmf(x, 3, 0.5)

plt.plot(x, fp, marker='o', linestyle='-', color='blue')
plt.xlabel('Número de sucessos')
plt.ylabel('Probabilidade')
plt.title(f'Distribuição Binomial (Eventos={n}, Probablidade={p})')
plt.xticks(x)
plt.grid(True)
plt.show()
```
