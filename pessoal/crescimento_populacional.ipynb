```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import numpy as np
from math import e, log
```

```python
pais = input("Qual país deseja analisar? ").upper()
tfr = float(input('Qual a taxa de nascimento por mulher? '))
pn = int(input('Qual a população atual? '))
```

```python
tfr_sub = 2.1
t = 1
r = log(tfr/tfr_sub) # Taxa de crescemento populacional (se < 2.1 é negativa - se > 2.1 é positiva e = 2.1 é nula)
x = np.arange(1, t + 10, 0.01) # Gerações
y = [(pn * e ** (r * t) / 1_000_000) for t in x] # Crescimento populacional
fig, ax = plt.subplots()
ax.plot(x, y)
ax.yaxis.set_major_formatter(ScalarFormatter())
ax.ticklabel_format(useOffset=False, style='plain', axis='y')
plt.title(f'VARIAÇÃO POPULACIONAL {pais}')
plt.xlabel('GERAÇÕES (25 anos)')
plt.ylabel('POPULAÇÃO (Em Milhões)')
plt.grid(True)
plt.show()
```

```python
# O Gráfico mostra que as imigrações para países como baixa natalidade, abaixo de 2.1 filhos/mulher é a solução para evitar uma desertificação dessas regiões populacionais.
```
