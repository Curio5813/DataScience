```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Brazilian Cities

# Context
# Brazil is the world's fifth-largest country by area, with 8.5 million square kilometer, and the fifth most populous, with over 208 million people.

# The Federative Republic of Brazil is composed of the union of the 26 states, the Federal District, and the 5,570 municipalities.

# This dataset is a compilation of several publicly available information about Brazilian Municipalities.

# Content
# The data was retrieved from several websites, as presented below. I processed and merged all data by city name, resulting in 79 fields for every city.
```

```python
dataset = pd.read_csv('brazil_cities.csv', sep=';')
print(dataset.shape)
dataset.head()
```

```python
data_pe_idhm_ed = dataset[(dataset['STATE'] == 'PE') & (dataset['IBGE_POP'] <= 50_000.00)]
print(data_pe_idhm_ed.shape)
data_pe_idhm_ed.head()
```

```python
idhm_edu = data_pe_idhm_ed['IDHM_Educacao']

media = np.mean(idhm_edu)
mediana = np.median(idhm_edu)
quartis = np.quantile(idhm_edu, [0, 0.25, 0.5, 0.75, 1])
desvio_padrao = np.std(idhm_edu, ddof=1)
variancia = np.var(idhm_edu, ddof=1)
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Quartis: {quartis}")
print(f"Desvio-Padrão: {desvio_padrao:.2f}")
print(f"Variância: {variancia:.2f}")
```

```python
# Teste de Normalidade - Histograma

plt.hist(idhm_edu, bins=20)
plt.title('IDHM Educacao Cidades PE <= 50.000 Habs.')
plt.show()
```

```python
# Teste de Normailidade - Gráfico Q-Q Plot

fig, ax = plt.subplots()
stats.probplot(idhm_edu, fit=True,   plot=ax)
plt.show()
```

```python
# Teste de Normailidade - Shapiro-Wilk
# p > 0.05 para Hipótese Nula

stats.shapiro(idhm_edu)
```

```python
# Cidade com o maior IDHM_Educacao
cidade_top = data_pe_idhm_ed.loc[data_pe_idhm_ed['IDHM_Educacao'].idxmax()]

print("Cidade com maior IDHM_Educacao:")
print(cidade_top[['CITY', 'STATE', 'IDHM_Educacao', 'IBGE_POP']])
```

```python
# Pelo Gráfico de Histograma e do Grafíco Q-Q Plot, o IDHM para Educação, no Estado de Pernambuco, para cidades com população igual ou menor que 50 mil habitantes estão virtualmente normalmente disbribuídos em torno da média, apesar de no Teste de Shapiro-Wilk o p < 0.05. Mas não há motivos para se rejeitar a hipótese nula. Muito provavelmete isto se deve a presença de outliers a presença de um IDHM para educação de 0.748 na cidade de Fernando de Noronha, arquipélogo turístico de Pernambuco com alto IDHM. A média de IDHM para Educação nas cidades do Estado de Pernambuco com população menor ou igual à 50 mil habitantes é de '0.48'.
```

```python
# A probabilidade de termos um IDHM para Educação < 0.40 pegando a cidades <= 50 mil habitantes é de:

prob = stats.norm.cdf(0.40, media, desvio_padrao)
print(f"A Probabilidade é de {prob * 100:.2f}%")
```

```python
# probabilidade de termo um IDHM para Educação >= 0.65 é de:

prob = stats.norm.sf(0.65, media, desvio_padrao)
print(f"A Probabilidade é de {prob * 100:.2f}%")
```

```python
# Conclusão: Cidade com uma população baixa no Estado de Pernambuco tem baixo IDHM para Educação. Isto é decorrência de uma política muito concentrada nas Federações e Municípios mais populosos. O que se verifica que não há, no Governo Estadual de Pernambuc,o uma distribuição suficiente de recursos para educação para os Municípios pouco populosos desse ente federado, devido, mais uma vez, à uma País fortemente centralizado, onde as políticas sociais partem do Governo Federal, distante da realidade das outras regiões do Estado brasileiro, através de repasse do FPM (Fundo de Participação dos Municípios).
```
