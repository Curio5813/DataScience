```python
import matplotlib.pyplot as plt
import numpy as np
from math import e
```

```python
x = np.arange(-1.1, 1.1, 0.00001)
y = [(e ** -i + e ** i) / 2 for i in x]
plt.plot(x, y)
plt.title("Curva Catenária")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

 # y = (e^(x) + e^(-x))/ 2
# A Curva Catenária é o lugar geométrico onde os ponto diametramente opostos na curva sofrem a mesma força gravitacional.
# É possível fazer um arco com está curva sem estruturas de vigas para sustentá-lo.
# É a curva formada qando deixamos uma corrente suspensa no ar, segurando apenas em suas extremidades.
# Essa curva é a única curva plana que se projetada em três dimenões forma um cubo.
```
