```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Numa cidade ocorrem em média 2 acidentes de carros por dia, qual a probabilidadde de ocorrerem mais de 3 acidentes de carros num dia?
# Usamos neste caso a Distribuição de Poisson, para valores discretos, pois estamos partindo da média de um evento acontecer, e não do espaço amostral, como acontece na Distribuição Binomial.

mu = 2
prob = 1 - stats.poisson.cdf(3, mu)
x = np.linspace(0, 50, 200)
y = stats.poisson.pmf(np.round(x), mu)

print(f"A probabilidade é de { prob * 100:.2f}%")
```

```python
# Gráfico da Distribuição de Poisson.

plt.plot(x, y)
plt.ylabel("Probability")
plt.xlabel("Numbers of Accidents")
plt.title("Poisson's Distribution")
plt.show()
```
