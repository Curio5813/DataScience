```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
N = 1000
r_min, r_max = 1e-35, 1e-33
mu_k = 2.0
t_array = np.linspace(0, 10, 200)
A = 1.2

# Geração dos nós fractais
alpha = 3.0
u = np.random.rand(N)
r = ((r_max**(1-alpha) - r_min**(1-alpha)) * u + r_min**(1-alpha))**(1/(1-alpha))

# Soma logarítmica
log_prod = np.sum(np.log((3*np.pi*r**3)/2.0))

# Trabalhar em log(E_S) para evitar overflow/underflow
log_E_S = log_prod + mu_k * (A**(-t_array/2.0) - 1.0)

# Normalização logarítmica (subtrai o primeiro valor)
log_E_S -= log_E_S[0]

# Converte para escala linear apenas para o gráfico
E_S = np.exp(log_E_S)

plt.figure(figsize=(8,5))
plt.plot(t_array, E_S, color='orange', linewidth=2)
plt.xlabel("Tempo (adimensional)")
plt.ylabel("E[S] / E[S](t=0)")
plt.yscale("log")
plt.title("Decaimento do Espaço-Tempo Quântico (modelo fractal)")
plt.grid(True, which="both", ls="--", lw=0.5)
plt.show()
```

```python
# Este modelo não reproduz o espaço-tempo fractal, mas apenas, os saltos quânticos quanto das interações das particulas elementares e o tempo como varíável discreta das observações, na função de onda de probabilidades. É apenas para mostrar que o modelo da Gravidade Quântica precisa de uma curva temporal para viabilizar as interações a nível elementares e extrapolar para níveis macroscópicos, para a realidade como vemos. O tempo é uma extrapolação do comportamento discreto da Mecânica Quântica.
```
