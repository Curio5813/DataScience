```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = int(input('Digite o valor do raio do círculo: '))
x = np.arange(-10, 10, 0.01)
x = np.linspace(-r, r, 500)   # só valores de -2 a 2
y1 = np.sqrt(r ** 2 - x ** 2)     # semicirculo superior
y2 = -np.sqrt(r ** 2 - x ** 2)    # semicirculo inferior
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Equação do Círculo')
plt.grid(True)
plt.show()

# Equação do círculo - Geometria Analítica
```
